```python
# this converts string into list
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

<!-- Now we would need to remove all the white spaces to prevent confusion in the model -->

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(1)
```

```python
# movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
# movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['title'].apply(lambda x: x.split())
```

```python
movies.head(1)
```

<!--  concatinating overview , genres , keywords, cast, crew into one "tag" column -->

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

<!-- now converting the lists into string -->

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

Stemming : to make words like love, loving, loved wrap in one word love. 

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

<!-- converting the text into lower case since its convinient -->

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

<!--  now we will do vectorization of words. so basically its going to count the occurence of a word in every movie. but it would ignore stop words. -->

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

<!--  now there's a problem that there are words like love, loving, loved etc which are treated differently and we dont want that. so we'll use "nltk" which is a natural language processing library which will help -->

to calculate distance between to movies. distance is inversely proportional to similarity. the more the distance the less the similaity.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Titanic')
```

```python
recommend('Batman')
```

```python
recommend('Thor')
```

```python
recommend('Harry Potter and the Philosopher\'s Stone')
```

```python
recommend('Harry Potter and the Chamber of Secrets')
```
